When sum(x^2) = sum(y^2), then the coefficient estimate same

```r
x = rnorm(100)
```

```r
y = x + rnorm(100)
```

```r
summary(lm(y~x+0))
```

```r
summary(lm(x~y+0))
```

#### When same coefficients are needed

```r
x = rnorm(100)
y = sample(x, 100)
```

```r
sum(x^2)
```

```r
sum(y^2)
```

```r
lm.fit = lm(y~x+0)
```

```r
lm.fit2 = lm(x~y+0)
```

```r
summary(lm.fit)
```

```r
summary(lm.fit2)
```

```r
plot(lm.fit)
```
