```r
lm.fit = lm(Sales ~ Price + Urban + US)
```

```r
summary(lm.fit)
```

```r
contrasts(Urban)
```

```r
contrasts(US)
```

```r
# Interpretation: Lower the price, higher the sales
# If US, higher sales\
# F statistic is high and p value is low, thus the prediction is statistically significant, 
# but R squared value is low, so not all the trends in data are predicted by these predictors
```

```r
# Given the same encoding for qualitative variables as above,
# Sales = 13.04 + (-0.054 * Price) + (-0.022 * UrbanYes) + (1.2 * USYes)
```

```r
# We can reject the null hypothesis for Price and US
```

```r
lm.fit2 = lm(Sales ~ Price + US)
```

```r
summary(lm.fit2)
```

```r
# Based on R squared and RSE, they both fit the data similarly
```

```r
confint(lm.fit2)
```

```r
plot(lm.fit2)
```

```r
# There are quite a bit of points on the leverage plot that exceed (p+1)/n = 0.0076, 
# thus indicating presence of high leverage points
```

```r
plot(predict(lm.fit2), rstudent(lm.fit2))
```

```r
# All the values are between -3 and 3
```
