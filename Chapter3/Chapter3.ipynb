```r
#  The basic syntax is lm(y∼x,data), where y is the response, x is the predictor, and
# data is the data set in which these two variables are kept
```

```r
lm.fit = lm(medv ~ lstat, data=Boston)
```

```r
lm.fit = lm(medv ~ lstat)
```

```r
lm.fit
```

```r
summary(lm.fit)
```

```r
names(lm.fit)
```

```r
# Coefficients
coef(lm.fit)
```

```r
# Confidence Intervals
confint(lm.fit)
```

```r
predict(lm.fit, data.frame(lstat=c(5, 10, 15)), interval='confidence')
```

```r
predict(lm.fit, data.frame(lstat=c(5, 10, 15)), interval='prediction')
```

```r
plot(lstat, medv, pch='+')
abline(lm.fit, lwd=3, col='red')
```

```r
# par(mfrow=c(2,2))
plot(lm.fit)
# Inference, the first plot tells that there is a non linear relationship in the data.
# If it were a straight line, it would have been a perfectly linear fit
```

```r
plot(predict(lm.fit), residuals(lm.fit))
```

```r
plot(predict(lm.fit), rstudent(lm.fit))
```

```r
plot(hatvalues(lm.fit))
# Leverage values
```

```r
which.max(hatvalues(lm.fit)) # returns the index of point with max leverage
```

#### Comparing 2 models

```r
lm.fit2 = lm(medv ~ lstat)
```

```r
summary(lm.fit2)
```

 The anova() function performs a hypothesis
test comparing the two models. The null hypothesis is that the two models
fit the data equally well, and the alternative hypothesis is that the full
model is superior. Here the F-statistic is 135 and the associated p-value is
virtually zero. This provides very clear evidence that the model containing
2the predictors lstat and lstat is far superior to the model that only
contains the predictor lstat. 

```r
anova(lm.fit, lm.fit2) # Stands for ANalysis Of VAriance table
```

```r
par(mfrow=c(2,2))
plot(lm.fit)
```
